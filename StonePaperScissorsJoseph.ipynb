```````````python
import random
import time
import os
def StonePaper():
    
    print()
    print("Welcome to Stone / Paper / Scissors")
    print()
 
    options=["Stone","Paper","Scissors"]
 
    for i in range(0,3):
        a=str(options[i]).lower()
        options[i]=a
 
    print("Lets Play the Game")
    print()
    print()    
    print("The Game begins...")
    time.sleep(1)
    print("___________________________________")
 
 
 
    
    def game():
        yourpoint=0
        comppoint=0
        i=0
        while i<5:
            print()
            print("Round  ",i+1,"Computer point= ",comppoint," &  Your Point= ",yourpoint)
            print()
            print("{0:<50}\n{0:^50}\n{0:>50}".format("Stone Paper Scissors"))
            print()
            hidden=random.choice(options)
#            print(hidden)
            answer=input("Play the game by entering your Answer here:  ")
            answer=answer.lower()
            print("Computer played---> ",hidden)
            time.sleep(1)
            
            if answer==hidden:
                print("Both have choose same options. Play again")
                time.sleep(1)
  #              os.system("cls")
            if answer not in options:
                print("********You may have miss spelled your answer.********")
                print("********Please enter correctly next time. Perfect okay?*******")
            elif answer!=hidden:
                if answer==options[0] and hidden==options[1] :
                    comppoint+=1
                    print("Computer scored")
                    print("You have ",yourpoint,"point. And computer have ",comppoint,"point." )
                    i+=1
                if answer==options[1] and hidden==options[2]:
                    comppoint+=1
                    print("Computer scored")
                    print("You have ",yourpoint,"point. And computer have ",comppoint,"point." )
                    i+=1
                if answer==options[2] and hidden==options[0]:
                    comppoint+=1
                    print("Computer scored")
                    print("You have ",yourpoint,"point. And computer have ",comppoint,"point." )
                    i+=1
                if answer==options[0] and hidden==options[2]:
                    yourpoint+=1
                    print("You scored")
                    print("You have ",yourpoint,"point. And computer have ",comppoint,"point." )
                    i+=1
                if answer==options[1] and hidden == options[0]:
                    yourpoint+=1
                    print("You scored")
                    print("You have ",yourpoint,"point. And computer have ",comppoint,"point." )
                    i+=1
                if answer==options[2] and hidden==options[1]:
                    yourpoint+=1
                    print("You scored")
                    print("You have ",yourpoint,"point. And computer have ",comppoint,"point." )
                    i+=1
                time.sleep(1)
 #               os.system("cls")
            print()
            print("___________________________________")
 
            
 
 
        if yourpoint>comppoint:
            time.sleep(1)
            print("You have win the game STONE PAPER SCISSORS with ",yourpoint,"points")
            print("Congratulations")
            print("         /|\   ")
            print("        / | \   ")
            print("       /  |  \   ")
            print("      /   |   \   ")
            print("      ``````````  ")
            print("     | O  *   O |   ")
            print("     |   \__/   |  ")
            print("     '\________/'   ")
            print()
            print()
            print("Game Over Best Of Luck")
            print()
        if comppoint>yourpoint:
            time.sleep(1)
            print()
            print("You failed in Game. your oponent has ",comppoint,"points")
            print("Better Luck next time")
            print()
            print()
    
    game()
```````````
